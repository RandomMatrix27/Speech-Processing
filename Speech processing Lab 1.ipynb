```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

file_path = "C:/Users/Sreyas/Desktop/speech1.ogg"

y, sr = librosa.load(file_path, sr=desired_sr)

y_reversed = np.flip(y)

plt.subplot(2, 1, 1)
librosa.display.waveshow(y, sr=sr)
plt.title('Original Waveform')

plt.subplot(2, 1, 2)
librosa.display.waveshow(y_reversed, sr=sr)
plt.title('Reversed Waveform')

plt.show()
```
